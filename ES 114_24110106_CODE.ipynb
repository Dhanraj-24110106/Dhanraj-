```python
def fibonacci(n):
    fib_sequence = []
    a, b = 0, 1
    while len(fib_sequence) < n:
        fib_sequence.append(a)
        a, b = b, a + b
    return fib_sequence

# Example usage
n = int(input())
fib_sequence = fibonacci(n)

# Create a number list for x-axis
number_list = list(range(1, n + 1))

# Plotting the graph
from matplotlib import pyplot as plt

plt.plot(number_list,fib_sequence)
plt.show()
```
